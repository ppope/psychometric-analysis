#Cluster analysis of 1000 psychometric profiles

##Goals
* Investigate clusters in the data using a variety of clustering methods.
* Estimate the density of the psychometric variables.

##Outline
* Used cosine similarity as primary metric. 
* DBSCAN
  * Requires two parameter choices
    * $\epsilon$ = "the maximum radius of neighborhood"
    * MinPts = "minimum number of points in an $\epsilon$-neighborhood of that point"
* Visualized clusters
  * Heatmaps
  * Pairwise plots of OCEAN scores
  * Densities of OCEAN scores by cluster
* Compute summary statistics on each cluster
  * Count
  * Mean
  * Standard deviation
  * Range
* Compute range of cosine scores for each cluster
* Made histogram of cosine scores.


```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.metrics.pairwise import pairwise_distances
from matplotlib import pyplot as plt
from matplotlib import cm 
```

```python
#Read in data
path = '/home/phil/deepmile/plenz/psychometrics/qn_scores_clean.txt'
Y = pd.read_table(path, header=None)
Y.columns = ['O','C','E','A','N']

#Calculate pairwise cosine distance.
# Recall: Distance = 1 - Similarity
X_dist = pairwise_distances(Y.values, metric='cosine')

# Compute DBSCAN
def calc_DBSCAN(dist_mat, epsilon, minimum_samples, print_opt=True):   
    db = DBSCAN(eps=epsilon, min_samples=minimum_samples, metric='precomputed').fit(dist_mat)
    labels = db.labels_
    # Number of clusters in labels, ignoring noise if present.
    # -1 denotes a noise data point, one not calculated to be in any cluster. 
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    if print_opt == True:
        if n_clusters_ < 2: #Not interested in results with 1 or 0 clusters
            return
        else: 
            print 'Epsilon:', epsilon, ', Minimum number of samples:', minimum_samples,', Estimated number of clusters:', n_clusters_
    else:
        return (labels, epsilon, minimum_samples)
```

#Last time:
###Visualized clustering from DBSCAN with heatmaps

Using the following parameter choices:

$\epsilon$ = 0.001, min_samples = 25

We obtained:

n_clusters_ = 3

Heatmaps using cosine distance for the above parameter choices were given below. Two versions are included, one with data points categorized as noise, and one without. 

```python

labels_and_params = calc_DBSCAN(X_dist, epsilon = 0.001, minimum_samples = 25, print_opt = False)

def make_heatmap(data, labels_and_params, drop_noise=True):
    Z = data.copy(deep=True) 
    Z['cluster_labels'] = labels_and_params[0]
    if drop_noise: Z = Z[Z.cluster_labels != -1] #Drop noise points
    Z = Z.sort(columns='cluster_labels') #Sort by cluster
    Z = Z.drop('cluster_labels', 1) #Drop the label again so it isn't plotted
    Z_dist = pairwise_distances(Z, metric='cosine') #Compute cosine distance
    heatmap = plt.imshow(Z_dist, cmap=cm.jet)
    #plt.title('DBSCAN Clusters (eps=0.001, min_samples = 25) Using Cosine Distance, Unclustered Points Not Included')
    title = 'DBSCAN clusters (' + 'eps = ' + str(labels_and_params[1]) + ', min_samples = ' + str(labels_and_params[2]) + ')' + ' using cosine distance'  
    if drop_noise: title += ' without noise'
    else: title += ' with noise'
    plt.title(title)
    plt.colorbar()

```

```python
make_heatmap(Y,labels_and_params, True)
```

#Pairwise plots of OCEAN scores

To visually inspect the clusters found with DBSCAN, pairwise scatterplots of each (unnormalized) OCEAN score were made.
Each OCEAN score is plotted against the other four below, along with a histogram for each score by cluster (including 'noise' cluster) .

For example, the first row from left to right contains,
1. histogram of the O scores by cluster
2. O vs. C 
3. O vs. E
4. O vs. A
5. O vs. N

The clusters are labeled as:

* -1  <-> noise points  <-> blue
* 0   <-> cluster zero  <-> green
* 1   <-> cluster one   <-> red
* 2  <-> cluster three <-> purple

As before, two versions are given: one with noise points, and one without.

Note: Each plots is a reflections of the plot across the diagonal. They show the same data but with axes flipped.
E.g. The (2,1) plot is O vs. C, and the (1,2) plot is C vs. O.

```python
#Resources for plotting
#http://stanford.edu/~mwaskom/software/seaborn/examples/scatterplot_matrix.html
#http://stackoverflow.com/questions/22943894/class-labels-in-pandas-scattermatrix
```

```python
def make_pairwise_plots(data, drop_noise=True):
    sns.set()
    if drop_noise: data = data[data.class_label != -1]
    data = data.sort('class_label', ascending=False)
    sns.pairplot(data, hue='class_label', size=2.5, vars=['O','C','E','A','N'], diag_kind='hist',
                 palette = {-1 : sns.color_palette()[0], 0 : sns.color_palette()[1],
                            1: sns.color_palette()[2], 2 : sns.color_palette()[3]})
```

```python
make_pairwise_plots(Z, drop_noise=False)
```

```python
make_pairwise_plots(Z, drop_noise=True)
```

#Compute summary statistics on each cluster and the entire population .

* Count of samples in each cluster (-1 denotes noise points)
* Mean

839 points were not clustered. 

Only 161/1000 = 16.1% were clustered. This means the clusters found tend to exclude most individuals in the population. The clusters may nonetheless may be of interest for identifying subpopulations with the data. 

Cluster 0 tended to have higher O and E scores, and lower C, A, and N scores than the population average.

Cluster 1 tended to have a higher N score, and lower O, C, E, and A scores than the population average. This cluster tended to be more neurotic.

Cluster 2 tended to have a higher O, C, E, and A scores, and lower N scores than the population average. This cluster tended to be more 'sociable'.

### Added histogram of cosine distances.

###Note: This histogram is of cosine distances, not with raw OCEAN values as above.

###Note: To make a histogram of cosine distances, must choose an arbitrary reference point from which distances to every other point are computed. Distances are always relative. The values used below refer to the distance between the reference point, choosen as the first point of the data, and all other points.

###Note: likewise for computing the range of cosine distances

```python
#Take first row (or column of distance matrix) to get cosine values relative to a reference point 
X_cosine = pd.DataFrame(X_dist[0])
X_cosine = X_cosine.join(labels_df)
X_cosine = X_cosine[X_cosine.class_label != -1]

#Slice data according to class label
X_cosine_0 = X_cosine[X_cosine.class_label == 0]
X_cosine_1 = X_cosine[X_cosine.class_label == 1]
X_cosine_2 = X_cosine[X_cosine.class_label == 2]

#Remove class label again
X_cosine_0 = X_cosine_0[0]
X_cosine_1 = X_cosine_1[0]
X_cosine_2 = X_cosine_2[0]

#Make histogram of cosine scores
sns.set()
X_cosine_0.hist(color=sns.color_palette()[1], bins=10)
X_cosine_1.hist(color=sns.color_palette()[2], bins=10)
X_cosine_2.hist(color=sns.color_palette()[3], bins=10)
plt.title('Histogram of cosine distances by cluster')
```
