```python
import pandas as pd
import numpy as np

from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.metrics.pairwise import pairwise_distances
from matplotlib import pyplot as plt
```

```python
#Read in data
path = '/home/phil/deepmile/plenz/psychometrics/qn_scores_clean.txt'
Y = pd.read_table(path, header=None)
Y.columns = ['O','C','E','A','N']
```

```python
#View first five rows of data 
Y.head()
```

```python
X = Y.as_matrix()
#Calculate pairwise cosine distance.
# Recall: Distance = 1 - Similarity
X_dist = pairwise_distances(X, metric='cosine')

# Compute DBSCAN
def calc_DBSCAN(dist_mat, epsilon, minimum_samples, print_opt=True):   
    db = DBSCAN(eps=epsilon, min_samples=minimum_samples, metric='precomputed').fit(dist_mat)
    labels = db.labels_
    # Number of clusters in labels, ignoring noise if present.
    # -1 denotes a noise data point, one not calculated to be in any cluster. 
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    if print_opt == True:
        if n_clusters_ < 2: #Not interested in results with 1 or 0 clusters
            return
        else: 
            print 'Epsilon:', epsilon, ', Minimum number of samples:', minimum_samples,', Estimated number of clusters:', n_clusters_
    else:
        return labels
```

```python
#Iterate over values for epsilon and min_samples to find interesting clusterings.
epsilion_list = np.linspace(0.001,1,1000)
min_samples_list = [1,5,10,15,25,50,100]
for epsilion in epsilion_list:
    for min_samples in min_samples_list:
        calc_DBSCAN(X_dist, epsilion, min_samples, True)
```
